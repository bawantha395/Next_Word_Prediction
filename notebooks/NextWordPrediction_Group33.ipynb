```python
with open("./news-headlines/ArticlesMarch2018.csv", 'r', encoding='utf-8') as file:
    news = file.read()
```

```python
news
```

```python
tokenizer = tf.keras.preprocessing.text.Tokenizer()
```

```python
tokenizer.fit_on_texts([news])
```

```python
tokenizer.get_config()
```

```python
tokenizer.word_index
```

```python
input_sequences = []
for sentence in news.split('\n'):
  #print(sentence)
  token_list = tokenizer.texts_to_sequences([sentence])[0]
  #print(token_list)
  for i in range(1, len(token_list)):
    sequence = token_list[:i+1]
    input_sequences.append(sequence)

```

```python
input_sequences[ :5]
```

```python
max_length = max([len(input_sequence) for input_sequence in input_sequences])
max_length
```

```python
# add padding
input_sequences = np.array(tf.keras.preprocessing.sequence.pad_sequences(input_sequences, maxlen=max_length, padding='pre'))
```

```python
input_sequences[1]
```

```python
# divide dataset into feature set (X) and labels (y)
X = input_sequences[:, :-1]
y = input_sequences[:, -1]
```

```python

X[0]
```

```python
y
```

```python
# number of words
num_classes = len(tokenizer.word_index)+1
num_classes
```

```python
def sparse_one_hot_encoding(y, num_classes):
    indices = tf.expand_dims(tf.range(0, len(y), dtype=tf.int64), 1)
    y_sparse = tf.SparseTensor(indices, y, dense_shape=[len(y), num_classes])
    return tf.sparse.to_dense(tf.sparse.reorder(y_sparse))

# Example usage
y_encoded_sparse = sparse_one_hot_encoding(y, num_classes)
```

```python
# Example of batch processing
batch_size = 1024  # Adjust based on your system's memory capacity

for i in range(0, len(y), batch_size):
    batch_labels = y[i:i + batch_size]
    batch_encoded = tf.keras.utils.to_categorical(batch_labels, num_classes=num_classes)
    # Now use batch_encoded for training in this batch
```

```python
# one-hot encoding labels
y = np.array(tf.keras.utils.to_categorical(y, num_classes=num_classes))
     
```

```python
y[1]
```

```python
# model building
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Embedding(num_classes, 80, input_length=max_length-1))
model.add(tf.keras.layers.LSTM(100))
model.add(tf.keras.layers.Dense(num_classes, activation='softmax'))
```

```python

model.summary()
```

```python

# compile the model
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
# model training
history = model.fit(X, y, epochs=100)
```

**Prediction Phase**


```python

input_text = "cricket begin"
for i in range(6):
    token_list = tokenizer.texts_to_sequences([input_text])[0]
    token_list = tf.keras.preprocessing.sequence.pad_sequences([token_list], maxlen=max_length-1, padding='pre')
    predicted = np.argmax(model.predict(token_list), axis=-1)
    output_word = ""
    for word, index in tokenizer.word_index.items():
        if index == predicted:
            output_word = word
            break
    input_text += " " + output_word

print(input_text)
```
